```python
import pandas as pd
```

```python
data = pd.read_csv('/content/per.csv' , encoding='utf-8')
```

```python
data.head(5)
```

```python
with open('/content/stopwords.txt' , encoding='utf8') as stopwords_file:
  stopwords = stopwords_file.readlines()
  stopwords= [line.replace('\n' , '') for line in stopwords]
  print(stopwords)
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
nltk_downloads = nltk.corpus.stopwords.words('english')
stopwords.extend(nltk_downloads)
len(stopwords)
```

```python
import hazm
stemmer = hazm.Stemmer()
stemmer.stem('کتاب ها')
```

```python
from hazm import word_tokenize
```

```python
dataset = pd.DataFrame(columns=('title_body' , 'category'))
for index , row in data.iterrows():
    title_body = row['Title'] + ' ' + row['Body']
    title_body_tokenizer = word_tokenize(title_body)
    title_body_tokenizer_filtred = [word for word in title_body_tokenizer if word not in stopwords]
    title_body_tokenizer_filtred_stemmed = [stemmer.stem(word) for word in title_body_tokenizer_filtred]

    dataset.loc[index] = {
        'title_body': ' '.join( title_body_tokenizer_filtred_stemmed),
        'category': row['Category2'].replace('\n', '')
    }
```

```python
dataset.head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer ()
```

```python
vectorizer.fit(dataset['title_body'])
```

```python
import pickle
with open ('vectorizer.jsdh' , 'wb') as f:
      pickle.dump(vectorizer , f)
```

```python
X = vectorizer.transform(dataset['title_body'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(dataset['category'])
```

```python
le.classes_
```

```python
import numpy as np
print(np.unique(dataset['category']))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y)
```

```python
from sklearn import svm
```

```python
svmc= svm.SVC()
svmc.fit(x_train , y_train)
```

```python
svmc.score(x_test ,y_test )
```

```python
from sklearn.metrics import classification_report , confusion_matrix
y_pred=svmc.predict(x_test)
```

```python
print(classification_report(y_test , y_pred))
```

```python
print(confusion_matrix(y_test , y_pred))
```
